```python
data=[]
for i in range(0,len(df)):
    result = df["Title"][i]+""+df["Uploader"][i]+""+df["Tags"][i]+df["FileType"][i]
    data.append(result)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vect = TfidfVectorizer()
data = vect.fit_transform(data)
```

```python
data.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Title = "Credit Card Fraud Detection"
Version = df[df["Title"]==Title]["Version"].values[0]
Version
```

```python
# Compute cosine similarity between items
cs = cosine_similarity(data, data)
```

```python
item_idx = list(df["Title"]).index(Title)
similarity_scores = list(enumerate(cs[item_idx]))
similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
similarity_scores[0][0]
```

```python
i = 0

for col in similarity_scores:
       
    print(df[df.index==col[0]]["Title"].values[0])
    i += 1
    if i >6:
        break
```

```python
def recommendaed_system(inputs):
    item_idx = list(df["Title"]).index(inputs)
    similarity_scores = list(enumerate(cs[item_idx]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    i = 0
    
    for col in similarity_scores:
           
        print(df[df.index==col[0]]["Title"].values[0])
        i += 1
        if i >6:
            break
```

```python
recommendaed_system("Bitcoin Historical Data")
```
